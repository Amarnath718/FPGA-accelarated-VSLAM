```python
import cv2
import torch
import matplotlib.pyplot as plt
from alike import ALike

# Load image
image = cv2.imread("chess.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Initialize model directly (Tiny version settings)
model = ALike(
    c1=8, c2=16, c3=32, c4=64,
    dim=64,
    single_head=True,
    radius=2,
    model_path="models/alike-t.pth",
    device="cpu",
    top_k=500
)

# Run detection
pred = model(image_rgb)
keypoints = pred['keypoints']

# Draw keypoints
for x, y in keypoints:
  cv2.circle(image_rgb, (int(x), int(y)), 2, (0, 255, 0), -1)

# Display
plt.figure(figsize=(8,8))
plt.imshow(image_rgb)
plt.title(f"Detected Keypoints: {len(keypoints)}")
plt.axis("off")
plt.show()
```

```python
pred = model(image_rgb)
```

```python
keypoints = pred['keypoints']
descriptors = pred['descriptors']

print("Number of keypoints:", len(keypoints))
print("Descriptor shape:", descriptors.shape)
```

```python
import time
start = time.time()

pred = model(image_rgb)

end = time.time()
execution_time = end - start

print("Execution Time (seconds):", execution_time)
```

```python
import numpy as np
times = []

for i in range(20):
    start = time.time()
    _ = model(image_rgb)
    end = time.time()
    times.append(end - start)

avg_time = np.mean(times)
fps = 1 / avg_time

print("Average Execution Time:", avg_time)
print("Average FPS:", fps)
```
